```python
import pandas as pd
```

```python
raw_titanic= pd.read_csv('data/train.csv', index_col='PassengerId')
raw_titanic.head()
```

```python
raw_titanic.info()
```

```python
raw_titanic.drop(['Name', 'Ticket', 'Fare', 'Cabin', 'Embarked'], axis=1, inplace=True)
```

```python
raw_titanic=pd.get_dummies(raw_titanic, columns=['Sex']) #Sex만 get_dummies
```

```python
raw_titanic['age_label'] = pd.cut(raw_titanic.Age, bins=[0, 2, 17, 65, 99], labels=['Baby', 'Child', 'Adult', 'Elderly']) #bins 0~2 baby, 2~17 child ... 라는 label 새로 만들어서 age_label이라는 새로운 col에 넣음
raw_titanic.head(10)
```

```python
raw_titanic.drop(['Age'], axis=1, inplace=True)
raw_titanic
```

```python
raw_titanic.fillna('Adult', inplace=True)
raw_titanic.head(10)
```

```python
titanic_data=pd.get_dummies(raw_titanic, columns=['age_label'])
titanic_data.head(10)
```

```python
X=titanic_data.iloc[:, 1:]
y=titanic_data.iloc[:, 0]
```

```python
X.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, shuffle=True)
```

```python
from tensorflow.keras import models, layers
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
model=models.Sequential()
model.add(layers.Dense(30, input_dim=9, activation='relu'))
model.add(layers.Dense(12, activation='relu'))
model.add(layers.Dense(8, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
history=model.fit(X, y, epochs=200, batch_size=5, verbose=False) #각 epoch 마다 검증
```

```python
score=model.evaluate(X_test, y_test) #test
print('Test Accuracy: %.4f'%(score[1])) #accuracy
```
