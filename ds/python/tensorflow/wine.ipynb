```python
#집값 예측 회귀 문제
from tensorflow.keras import models, layers
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical

import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
import numpy as np
```

```python
white= pd.read_csv("data/winequality-white.csv", delimiter= ';') #header=None(하면 header도 data로 읽음)
```

```python
white.head()
```

```python
new_row = pd.DataFrame({'white': [1]})
white= white.append(new_row, ignore_index=True)
```

```python
white['white'] = 1
```

```python
white
```

```python
red=pd.read_csv('data/winequality-red.csv', delimiter=';')
```

```python
red.head()
```

```python
new_row1 = pd.DataFrame({'white': [0]})
red= red.append(new_row1, ignore_index=True)
```

```python
red['white'] = 0
```

```python
red
```

```python
df = pd.concat([white, red])
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum().sort_values(ascending=False).head(20)
```

```python
df.dropna(inplace=True)
```

```python
df1=df.drop(['quality'], axis=1)
```

```python
df1.head()
```

```python
X=df1.iloc[:,:]
y=df['quality'].values
```

```python
X.shape
```

```python
y.shape
```

```python
y=to_categorical(y)
```

```python
y
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, shuffle=True)
```

```python
from tensorflow.keras import models, layers

model=models.Sequential()
model.add(layers.Dense(200, input_dim=12, activation='relu'))
model.add(layers.Dense(100, activation='relu'))
model.add(layers.Dense(80, activation='relu'))
model.add(layers.Dense(40, activation='relu'))
model.add(layers.Dense(20, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy']) #다중분류 손실함수
```

```python
history=model.fit(X_train, y_train, epochs=500, batch_size=100, verbose= True)
```

```python
score=model.evaluate(X_test, y_test) #test
print('Test Accuracy: %.4f'%(score[1])) #accuracy
```
