```python
import pandas as pd
red_wine = pd.read_csv('data/winequality-red.csv', sep=';')
red_wine.head()
```

```python
white_wine = pd.read_csv('data/winequality-white.csv', sep=';')
white_wine.head()
```

```python
red_wine.insert(11, 'class', 0)
white_wine.insert(11, 'class', 1)
```

```python
wine_df = pd.concat([red_wine, white_wine], ignore_index=True)
wine_df
```

```python
from sklearn.preprocessing import MinMaxScaler

mm_scaler = MinMaxScaler()
wine_df[['fixed acidity', 'residual sugar', 'free sulfur dioxide', 
         'total sulfur dioxide', 'pH', 'alcohol']] = \
mm_scaler.fit_transform(wine_df[['fixed acidity', 'residual sugar', 
                                 'free sulfur dioxide', 
                                 'total sulfur dioxide', 'pH', 'alcohol']])
wine_df
```

```python
X = wine_df.iloc[:, :12]
y = wine_df.iloc[:, 12].values
```

```python
from tensorflow.keras.utils import to_categorical

y = to_categorical(y, 11)
y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True)
```

```python
from tensorflow.keras import models, layers, optimizers

model = models.Sequential()
model.add(layers.Dense(24, input_dim=12, activation='relu'))
model.add(layers.Dense(24, activation='relu'))
model.add(layers.Dense(11, activation='softmax'))
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer=optimizers.Adam(learning_rate=0.05), #learning rate 0.05
              metrics=['accuracy'])
h = model.fit(X_train, y_train, epochs=60, batch_size=300)
```

```python
score = model.evaluate(X_test, y_test)
print('Test Accuracy:%.4f'% (score[1]))
```
