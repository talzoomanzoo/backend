```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X1=np.arange(-1.0, 1.0, 0.2)
X2=np.arange(-1.0, 1.0, 0.2)
```

```python
Z=np.zeros((10, 10))
```

```python
#W=np.array([2.5, 3.0])
#W=np.array([-2.5, -3.0])
W=np.array([0, 3.0])
#bias=np.array([0.1])
#bias=np.array([-2.0])
bias=np.array([2.0])
```

```python
for i in range(10):
    for j in range(10):
        X=np.array([X1[i], X2[j]])
        u=np.dot(X, W.T) + bias
        y=1/(1+np.exp(-u))
        Z[j][i]=y
```

```python
plt.imshow(Z, 'gray', vmin=0.0, vmax=1.0)
plt.colorbar()
plt.show()
```
