```python
import requests
from bs4 import BeautifulSoup as bs
#chrome driver XXX.XXXX 크롬버전 확인 후 맞는 버전으로 다운로드
#conda activate p39-env
#conda install selenium=4 #아나콘다가 지원하지 않음 따라서 별도의 채널
# https://anaconda.org/conda-forge/selenium 확인
# conda install -c conda-forge selenium=4
# selenium 4.9.1 있는 것 확인 후 proceed
from selenium import webdriver #팝업창 자동으로 닫아주기 필요
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome(service = Service('D:\source\web_scrap\driver\chromedriver.exe'))
```

```python
driver.get('https://vibe.naver.com/chart/total')
```

```python
xpath = """//*[@id="app"]/div[2]/div/div/a[2]""" #닫기창 xpath 복붙
driver.find_element(By.XPATH, xpath).click()
```

```python
soup = bs(driver.page_source, 'lxml')
print(soup)
```

```python
driver.close()
```

```python
def rank_list(driver_directory, attribute):
	driver_list = soup.select(driver_directory)
	make_list = []
	for i in range(len(driver_list)):
		if attribute == 'class' or attribute == 'id':
			make_list.append(driver_list[i].text)
		else:
			make_list.append(driver_list[i].get(attribute))
	
	return make_list

def rank_print(*args):
    for i, v in enumerate(zip(*args)):
        print(i+1, end = ' ')
        for j in range(len(v)):
            print(v[j], end = ' ')
        print('\n')
```
